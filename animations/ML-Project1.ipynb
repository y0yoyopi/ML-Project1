```python
%%manim -ql HeightWeightTable
from manim import *

class HeightWeightTable(Scene):
    def construct(self):
        np.random.seed(67)
        heights = np.random.uniform(150, 190, 10)
        weights = 0.8 * heights - 30 + np.random.normal(0, 12, 10)
        
        heights_rounded = np.round(heights, 1)
        weights_rounded = np.round(weights, 1)
        
        table_data = []
        for i, (h, w) in enumerate(zip(heights_rounded, weights_rounded), 1):
            table_data.append([f"{i}", f"{h:.1f}", f"{w:.1f}"])
        
        col_titles = ["# Persona", "Altura (cm)", "Peso (kg)"]
        colors = [RED, BLUE, GREEN]  # Colores para cada columna

        table = Table(
            table_data,
            col_labels=[
                Text(title, font_size=45, color=color)
                for title, color in zip(col_titles, colors)
            ],
            include_outer_lines=True,
            line_config={"stroke_width": 2},
            element_to_mobject_config={"font_size": 45}
        ).scale(0.45).move_to(ORIGIN)
        
        for i in range(1, 11):  
            table.get_entries((i, 0)).set_color(WHITE)
                
        for entry in table.get_entries():
            entry.set_opacity(0)
        self.play(
            AnimationGroup(
                *[entry.animate.set_opacity(1) for entry in table.get_entries()],
                lag_ratio=0.03
            ),
            run_time=0.5
        )
        self.play(
            Create(table.get_horizontal_lines()),
            Create(table.get_vertical_lines()),
            run_time=0.5
        )
        self.wait(2)
```

```python
%%manim -ql HeightWeightScatter #cambiar para más resolución
from manim import *
import numpy as np

class HeightWeightScatter(Scene):
    def construct(self):
        np.random.seed(67)
        heights = np.random.uniform(150, 190, 10)
        weights = 0.8 * heights - 30 + np.random.normal(0, 12, 10)

        axes = Axes(
            x_range=[140, 200, 10],
            y_range=[50, 140, 10],  
            x_length=5,
            y_length=5,
            axis_config={
                "include_numbers": True,
                "font_size": 20,
                "tip_width": 0,
                "tip_height": 0,
                "stroke_width": 2,
            },
            y_axis_config={"tip_width": 0.1, "tip_height": 0.2},
            x_axis_config={"tip_width": 0.1, "tip_height": 0.2}
        ).scale_to_fit_width(config.frame_width * 0.47).shift(UP * 0.5).shift(RIGHT * 0.2)

        x_label = Text("Altura (cm)", font_size=24).next_to(axes.x_axis, DOWN, buff=0.3)
        y_label = Text("Peso (kg)", font_size=24).next_to(axes.y_axis, LEFT, buff=0.3)

        dots = VGroup(*[
            Dot(axes.coords_to_point(h, w), color=BLUE)
            for h, w in zip(heights, weights)
        ])

        self.play(
            FadeIn(axes),  
            FadeIn(x_label),
            FadeIn(y_label),
            run_time=1
        )
        self.play(
            AnimationGroup(
                *[DrawBorderThenFill(d) for d in dots],
                lag_ratio=0  
            ),
            run_time=1
        )
        #2s
        self.wait(3)

        indices = [9,2] 
        lines_and_texts = []  
        
        for idx in indices:
            dot = dots[idx]
            original_color = dot.get_color()
            
            self.play(
                dot.animate.set_color(YELLOW).scale(1.5),
                run_time=0.3
            )
            
            h, w = heights[idx], weights[idx]
            point = axes.coords_to_point(h, w)
            
            h_line = DashedLine(
                start=point,
                end=axes.coords_to_point(140, w),
                color=YELLOW,
                dash_length=0.05,
                dashed_ratio=0.5
            )
            
            v_line = DashedLine(
                start=point,
                end=axes.coords_to_point(h, 50),
                color=YELLOW,
                dash_length=0.05,
                dashed_ratio=0.5
            )
            
            texto = Text(
                f"Altura: {h:.1f} cm\nPeso: {w:.1f} kg",
                font_size=20,
                color=YELLOW
            ).next_to(dot, UP + RIGHT, buff=0.1)
            
            lines_and_texts.extend([h_line, v_line, texto])
            
            self.play(Create(h_line), Create(v_line), run_time=0.5)
            self.play(FadeIn(texto), run_time=0.3)
            
            self.play(
                dot.animate.set_color(original_color).scale(1/1.5),
                run_time=0.4
            )

            self.wait(5)
        
        self.wait(1)
        
        self.play(
            *[FadeOut(element) for element in lines_and_texts],
            run_time=0.8
        )
        

        
        #2s

        
        
        
        var_text = VGroup(
            Text("X: Variable Independiente (Altura)", font_size=22, color=YELLOW),
            Text("Y: Variable Dependiente (Peso)", font_size=22, color=YELLOW),
        ).arrange(DOWN, aligned_edge=LEFT, buff=0.2)
        
        var_text.next_to(axes, RIGHT, buff=1.5).shift(DOWN * 1.3).shift(LEFT * 2.7)
        
        self.play(
            LaggedStart(
                *[Write(text) for text in var_text],
                lag_ratio=0.2
            ),
            run_time=2
        )

        self.wait(1)  
        self.play(FadeOut(var_text), run_time=0.5) 
        self.wait(0.5) 
        
        # X = 170 cm hasta punto medio de la gráfica
        x_value = 170
        y_mid = (axes.y_range[0] + axes.y_range[1]) / 2
        y_mid = (50 + 140) / 2  # 95
        
        x_line = DashedLine(
            start=axes.coords_to_point(x_value, 50),  # Desde abajo
            end=axes.coords_to_point(x_value, y_mid), # Hasta la mitad de Y
            color=GREEN,
            dash_length=0.05,
            dashed_ratio=0.5,
            stroke_width=3
        )
        
        altura_text = Text(f"Altura: {x_value} cm", font_size=24, color=GREEN)
        peso_estimado = 0.8 * x_value - 30
        peso_text = Text(f"Peso: ? ", font_size=24, color=GREEN)
        
        info_text = VGroup(altura_text, peso_text).arrange(DOWN, aligned_edge=LEFT, buff=0.2)
        
        info_text.next_to(axes, RIGHT, buff=1.5).shift(DOWN * 1.3).shift(LEFT * 2.3)
        
        self.play(
            Create(x_line, run_time=1.5),
            LaggedStart(
                Write(altura_text),
                Write(peso_text),
                lag_ratio=0.3
            ),
            run_time=2
        )
        self.wait(2)


        self.play(FadeOut(altura_text, peso_text, x_line), run_time=0.5) 

        # CONCEPTO REGRESIÓN LINEAL
        reg_text = Text(
            "Regresión Lineal: busca la mejor\nlínea recta que se ajuste a los\ndatos para hacer predicciones",
            font_size=20,
            color=YELLOW,
            line_spacing=0.8
        ).next_to(axes, RIGHT, buff=1.5).shift(DOWN * 1.3).shift(LEFT * 2.7)
        
        self.play(Write(reg_text), run_time=1.5)
        self.wait(2) 
        
        x = heights
        y = weights
        n = len(x)
        
        m = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2)
        b = (np.sum(y) - m*np.sum(x)) / n
        
        # DIBUJAR LÍNEA DE REGRESIÓN
        x_min, x_max = 140, 200
        y_min = m * x_min + b
        y_max = m * x_max + b
        
        reg_line = Line(
            start=axes.coords_to_point(x_min, y_min),
            end=axes.coords_to_point(x_max, y_max),
            color=RED,
            stroke_width=4
        )
        
        self.play(Create(reg_line), run_time=2)
        self.wait(2)  # Mostrar línea 2 segundos
        
        self.play(FadeOut(reg_text), run_time=0.8)
        
        self.wait(1)




        
        # ALTURA 170 Y PESO EXACTO
        x_pred = 170
        y_pred = m * x_pred + b 
        
        vertical_line = DashedLine(
            start=axes.coords_to_point(x_pred, 50),  # Desde eje X
            end=axes.coords_to_point(x_pred, y_pred),  # Hasta línea de regresión
            color=GREEN,
            dash_length=0.05,
            dashed_ratio=0.5,
            stroke_width=3
        )
        
        prediction_point = Dot(
            axes.coords_to_point(x_pred, y_pred),
            color=GREEN,
            radius=0.08
        )
        
        horizontal_line = DashedLine(
            start=axes.coords_to_point(x_pred, y_pred),  # Desde el punto
            end=axes.coords_to_point(140, y_pred),  # Hasta eje Y
            color=GREEN,
            dash_length=0.05,
            dashed_ratio=0.5,
            stroke_width=3
        )
        
        prediction_text1 = Text(
            f"Altura: {x_pred} cm", font_size=24, color=GREEN
        ).next_to(axes, RIGHT, buff=1.5).shift(DOWN * 1.1).shift(LEFT * 2.3)

        prediction_text2 = Text(
            f"Peso: {y_pred:.1f} kg", font_size=24, color=GREEN
        ).next_to(axes, RIGHT, buff=1.5).shift(DOWN * 1.6).shift(LEFT * 2.3)
        
        
        self.play(Create(vertical_line), run_time=1)
        self.wait(0.3)
        self.play(FadeIn(prediction_point), run_time=0.5)
        self.wait(0.3)
        self.play(Write(prediction_text1), run_time=1)
        self.wait(0.3)
        self.play(Create(horizontal_line), run_time=1)
        self.wait(0.3)
        self.play(Write(prediction_text2), run_time=1)
        
        self.wait(2) 
        
        self.play(
            FadeOut(vertical_line),
            FadeOut(prediction_point),
            FadeOut(horizontal_line),
            FadeOut(prediction_text1),
            FadeOut(prediction_text2),
            run_time=0.8
        )
        
        self.wait(1)

        # línea 2 que pasa por (140, 68) y (188, 111)
        x1_2, y1_2 = 140, 68
        x2_2, y2_2 = 188, 111
        
        m2 = (y2_2 - y1_2) / (x2_2 - x1_2)  # = (111 - 68) / (188 - 140) = 43/48 ≈ 0.8958
        b2 = y1_2 - m2 * x1_2  # = 68 - 0.8958*140 ≈ 68 - 125.42 = -57.42
        
        x_min2, x_max2 = 140, 200
        y_min2 = m2 * x_min2 + b2
        y_max2 = m2 * x_max2 + b2 
        
        reg_line2 = Line(
            start=axes.coords_to_point(x_min2, y_min2),
            end=axes.coords_to_point(x_max2, y_max2),
            color=ORANGE,
            stroke_width=4
        )
        
        # línea 3 que pasa por (140, 50) y (183, 122)
        x1_3, y1_3 = 140, 50
        x2_3, y2_3 = 183, 122
        
        m3_original = (y2_3 - y1_3) / (x2_3 - x1_3) 
        b3_original = y1_3 - m3_original * x1_3
        
        x_min3, x_max3 = 140, 200
        y_min3 = m3_original * x_min3 + b3_original
        y_max3 = m3_original * x_max3 + b3_original 

        reg_line3 = Line(
            start=axes.coords_to_point(140, m3_original*140 + b3_original),
            end=axes.coords_to_point(200, m3_original*200 + b3_original),
            color=GREEN,
            stroke_width=4
        )

        #cuál es la más óptima?

        self.play(FadeOut(reg_line), run_time=0.1)
        self.play(Create(reg_line2), run_time=0.2)
        self.wait(0.4)
        self.play(FadeOut(reg_line2), run_time=0.1)
        self.play(Create(reg_line3), run_time=0.2)
        self.wait(1.5)

        self.play(Create(reg_line),Create(reg_line2),Create(reg_line3), run_time=0.5)
        
        self.wait(6)
       

        
    
```

```python
%%manim -ql DiagonalLineDemo
from manim import *
import numpy as np

class DiagonalLineDemo(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0, 10, 2],
            y_range=[0, 10, 2],  
            x_length=5,
            y_length=5,
            axis_config={
                    "include_numbers": False,
                "font_size": 20,
                "tip_width": 0,
                "tip_height": 0,
                "stroke_width": 2,
            },
            y_axis_config={"tip_width": 0.1, "tip_height": 0.2},
            x_axis_config={"tip_width": 0.1, "tip_height": 0.2}
        ).scale_to_fit_width(config.frame_width * 0.47).shift(UP * 0.3).shift(LEFT * 0.2)

        x_label = Text("X", font_size=24).next_to(axes.x_axis, DOWN, buff=0.3)
        y_label = Text("Y", font_size=24).next_to(axes.y_axis, LEFT, buff=0.3)

        self.play(FadeIn(axes), FadeIn(x_label), FadeIn(y_label), run_time=1)
        self.wait(0.5)

        # línea diagonal perfecta
        m4_original = 1.0 
        b4_original = 0.0 
        
        x_min, x_max = 0, 10
        y_min = m4_original * x_min + b4_original  
        y_max = m4_original * x_max + b4_original  
        
        diag_line = Line(
            start=axes.coords_to_point(x_min, y_min),
            end=axes.coords_to_point(x_max, y_max),
            color=BLUE,
            stroke_width=4
        )

        # fórmula y = mx+b
        formula = MathTex("y", "= ", "m", "x", " + ", "b", font_size=60, color=YELLOW)
        formula.next_to(axes, RIGHT, buff=1.5).shift(DOWN * 0.3).shift(LEFT * 2.1)
        
        self.play(
            Create(diag_line, run_time=1.5),
            
        )
        self.wait(1.5)

        self.play(Write(formula, run_time=1),run_time=1)

        # encerrar "y" 
        y_box = SurroundingRectangle(formula[0], color=WHITE, buff=0.1)
        output_text = Text("output", font_size=24, color=WHITE).next_to(y_box, DOWN, buff=0.1)
        
        self.play(Create(y_box), Write(output_text), run_time=1)
        self.wait(1)
        self.play(FadeOut(y_box), FadeOut(output_text), run_time=0.5)
        
        
        # encerrar "x"
        x_box = SurroundingRectangle(formula[3], color=WHITE, buff=0.1)
        input_text = Text("input", font_size=24, color=WHITE).next_to(x_box, DOWN, buff=0.1)
        
        self.play(Create(x_box), Write(input_text), run_time=1)
        self.wait(1)
        self.play(FadeOut(x_box), FadeOut(input_text), run_time=0.5)
        

        m_values = [0.2, 1.0, 2.0]
        m_duration = 2.0 / len(m_values)  
        
        m_highlight = formula[2].copy()  # "m" está en índice 2
        m_highlight.set_color(RED).scale(1.3)

        # mover pendiente m
        for new_m in m_values:
            y1 = new_m * x_min  
            y2 = new_m * x_max 
            
            if y2 > 10:
                x_adj = 10 / new_m if new_m > 0 else x_max
                new_start = axes.coords_to_point(x_min, y1)
                new_end = axes.coords_to_point(min(x_adj, x_max), 10)
            else:
                new_start = axes.coords_to_point(x_min, y1)
                new_end = axes.coords_to_point(x_max, y2)
            
            # resaltar m en rojo
            self.play(
                Transform(formula[2], m_highlight),  # Resaltar m
                diag_line.animate.put_start_and_end_on(new_start, new_end),
                run_time=m_duration * 0.8  # 80% del tiempo para la animación
            )
            
            self.wait(m_duration * 0.1)  # 10% pausa
            
            temp_normal = formula[2].copy().set_color(YELLOW).scale(1/1.3)
            self.play(
                Transform(formula[2], temp_normal),
                run_time=m_duration * 0.1  # 10% para quitar resaltado
            )
        
        # volver ecuación original
        final_start = axes.coords_to_point(x_min, m4_original * x_min)
        final_end = axes.coords_to_point(x_max, m4_original * x_max)
        self.play(
            diag_line.animate.put_start_and_end_on(final_start, final_end),
            run_time=0.5
        )
        self.wait(0.5)

        # variar b4 con m1 es´tatico
        b_values = [-3 , 3, 0]
        b_duration = 2.0 / len(b_values)  
        
        b_highlight = formula[5].copy() 
        b_highlight.set_color(RED).scale(1.3)
        
        for new_b in b_values:
            x1 = max(0, -new_b)
            x2 = min(10, 10 - new_b)
            
            if x1 < x2:
                y1 = x1 + new_b
                y2 = x2 + new_b
                
                start_point = axes.coords_to_point(x1, y1)
                end_point = axes.coords_to_point(x2, y2)
                
                # resaltar b en rojo 
                self.play(
                    Transform(formula[5], b_highlight),
                    diag_line.animate.put_start_and_end_on(start_point, end_point),
                    run_time=b_duration * 1 
                )
            
            self.wait(b_duration * 0.15) 
            
            temp_normal = formula[5].copy().set_color(YELLOW).scale(1/1.3)
            self.play(
                Transform(formula[5], temp_normal),
                run_time=b_duration * 0.15
            )
        
        # volver a diagonal perfecta 
        final_diag_start = axes.coords_to_point(0, 0)
        final_diag_end = axes.coords_to_point(10, 10)
        self.play(
            diag_line.animate.put_start_and_end_on(final_diag_start, final_diag_end),
            run_time=0.5
        )
        
        # restaurar colores fórmula
        self.play(
            formula[2].animate.set_color(YELLOW).scale(1),
            formula[5].animate.set_color(YELLOW).scale(1),
            run_time=0.5
        )

        self.wait(1)
```

```python
%%manim -ql HeightWeightScatter
from manim import *
import numpy as np

class HeightWeightScatter(Scene):
    def construct(self):
        np.random.seed(67)
        heights = np.random.uniform(150, 190, 10)
        weights = 0.8 * heights - 30 + np.random.normal(0, 12, 10)

        axes = Axes(
            x_range=[140, 200, 10],
            y_range=[50, 140, 10],  
            x_length=5,
            y_length=5,
            axis_config={
                "include_numbers": True,
                "font_size": 20,
                "tip_width": 0,
                "tip_height": 0,
                "stroke_width": 2,
            },
            y_axis_config={"tip_width": 0.1, "tip_height": 0.2},
            x_axis_config={"tip_width": 0.1, "tip_height": 0.2}
        ).scale_to_fit_width(config.frame_width * 0.47).shift(UP * 0.5).shift(LEFT * 0.6)

        x_label = Text("Altura (cm)", font_size=24).next_to(axes.x_axis, DOWN, buff=0.3)
        y_label = Text("Peso (kg)", font_size=24).next_to(axes.y_axis, LEFT, buff=0.3)

        dots = VGroup(*[
            Dot(axes.coords_to_point(h, w), color=BLUE)
            for h, w in zip(heights, weights)
        ])

        self.play(
            FadeIn(axes),
            FadeIn(x_label),
            FadeIn(y_label),
            AnimationGroup(
                *[DrawBorderThenFill(d) for d in dots],
                lag_ratio=0  
            ),
            run_time=0.5
        )

        # regresión lineal usando least squares
        x = heights
        y = weights
        n = len(x)
        
        m = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2)
        b = (np.sum(y) - m*np.sum(x)) / n
        
        x_min, x_max = 140, 200
        y_min = m * x_min + b
        y_max = m * x_max + b
        
        reg_line = Line(
            start=axes.coords_to_point(x_min, y_min),
            end=axes.coords_to_point(x_max, y_max),
            color=RED,
            stroke_width=4
        )
        
        self.play(Create(reg_line), run_time=0.5)
        self.wait(1.5)

        # residuales
        residuals = VGroup()
        residual_lines = [] 
        
        for i, (h, w) in enumerate(zip(heights, weights)):
            y_pred = m * h + b
            
            # Crear línea residual, diferencia
            point_real = axes.coords_to_point(h, w)
            point_pred = axes.coords_to_point(h, y_pred)
            
            residual_line = Line(
                start=point_real,
                end=point_pred,
                color=YELLOW,
                stroke_width=2
            )
            
            residual_lines.append(residual_line)
            residuals.add(residual_line)
            

        self.play(
            LaggedStart(
                *[Create(line) for line in residual_lines],
                lag_ratio=0.1
            ),
            run_time=1.5
        )
        self.wait(4)


        # crear cuadrados
        squares = VGroup()
        square_anims = []

        
        for line, (h, w) in zip(residual_lines, zip(heights, weights)):
            length = line.get_length()
        
            y_pred = m*h + b
            residual = w - y_pred
        
            # decidir dirección según signo, para que no se interpongan los cuadrados
            direction = LEFT if residual > 0 else RIGHT
        
            sq = Square(
                side_length=length,
                color=YELLOW,
                stroke_width=2,
                fill_color=YELLOW,
                fill_opacity=0.3
            )
        
            sq.move_to(line.get_center())
            sq.shift(direction * (length / 2))
        
            squares.add(sq)
            square_anims.append(ReplacementTransform(line.copy(), sq))
        
        self.play(
            LaggedStart(
                *square_anims,
                lag_ratio=0.1
            ),
            run_time=2
        )
        self.wait(2)

        # SSE
        areas = [(sq.side_length)**2 for sq in squares]
        SSE = sum(areas)
        big_side = np.sqrt(SSE)
        
        #cuadrado grande a la derecha
        big_square = Square(
            side_length=big_side,
            color=YELLOW,
            stroke_width=3,
            fill_color=YELLOW,
            fill_opacity=0.3
        )
        
        big_square.next_to(axes, RIGHT, buff=1.5)
        
        # 3. cuadrados pequeños -> cuadrado grande
        self.play(
            LaggedStart(
                *[Transform(sq, big_square.copy()) for sq in squares],
                lag_ratio=0.1
            ),
            run_time=2
        )
        
        self.wait(1)
        
        # desaparecen los cuadrados pequeños y dejamos solo el cuadrado grande
        self.add(big_square)

        self.play(
            *[FadeOut(sq) for sq in squares],
            run_time=0.5
        )

        
        self.wait(1)
        
        # agregar texto SSE
        sse_text = Text("Sum of Squares", font_size=36).move_to(big_square.get_center())
        self.play(Write(sse_text), run_time=1.5)
        self.wait(2)

        self.play(
            FadeOut(sse_text),
            FadeOut(big_square),
            runtime = 0.5)
        
        self.wait(2)




        

        re_squares = VGroup()
        re_lines = []
        b_tracker = ValueTracker(b)

        for (h, w) in zip(heights, weights):
            y_pred = m*h + b
            point_real = axes.coords_to_point(h, w)
            point_pred = axes.coords_to_point(h, y_pred)

            # residual line
            re_line = Line(
                point_real,
                point_pred,
                color=YELLOW,
                stroke_width=2
            )
            re_lines.append(re_line)

            length = re_line.get_length()
            residual = w - y_pred
            direction = LEFT if residual > 0 else RIGHT

            sq = Square(
                side_length=length,
                color=YELLOW,
                stroke_width=2,
                fill_color=YELLOW,
                fill_opacity=0.3
            ).move_to(re_line.get_center()).shift(direction*(length/2))

            re_squares.add(sq)

        self.play(
            *[Create(l) for l in re_lines],
            *[Create(sq) for sq in re_squares],
            run_time=1.5
        )

        self.play(
            LaggedStart(
                *[FadeOut(line) for line in residual_lines],
                lag_ratio=0.1
            ),
            run_time=0.1
        )

        self.wait(1)


        b_tracker = ValueTracker(b)

        # updater para animación de girar la regresión lineal
        def updater(mob):
            new_b = b_tracker.get_value()

            # actualizar recta
            y_min2 = m * x_min + new_b
            y_max2 = m * x_max + new_b
            reg_line.put_start_and_end_on(
                axes.coords_to_point(x_min, y_min2),
                axes.coords_to_point(x_max, y_max2)
            )

            # actualizar residuales + cuadrados
            for i, (h, w) in enumerate(zip(heights, weights)):
                y_pred2 = m*h + new_b
                point_real = axes.coords_to_point(h, w)
                point_pred2 = axes.coords_to_point(h, y_pred2)

                line = re_lines[i]
                square = re_squares[i]

                # línea residual
                line.put_start_and_end_on(point_real, point_pred2)

                # cuadrado
                length2 = line.get_length()
                residual2 = w - y_pred2
                direction2 = LEFT if residual2 > 0 else RIGHT

                square.become(
                    Square(
                        side_length=length2,
                        color=YELLOW,
                        stroke_width=2,
                        fill_color=YELLOW,
                        fill_opacity=0.3
                    ).move_to(line.get_center()).shift(direction2*(length2/2))
                )

        # activar updater
        reg_line.add_updater(updater)

        # hasta b+25
        self.play(b_tracker.animate.set_value(b + 25), run_time=3)
        self.wait(0.5)

        #hasta b-8, justo origen
        self.play(b_tracker.animate.set_value(b - 8), run_time=3)
        self.wait(0.5)

        # volver al b original
        self.play(b_tracker.animate.set_value(b), run_time=3)
        self.wait(1)

        # detener updater y dejar recta fija
        reg_line.remove_updater(updater)

        self.play(
            LaggedStart(
                *[FadeOut(sq) for sq in re_squares],
                *[FadeOut(l) for l in re_lines],
                lag_ratio=0.1
            ),
            run_time=0.2
        )

        self.wait(5)
```

```python

def build_features(x, degree=1):
    x = np.atleast_2d(x).T if x.ndim == 1 else x
    features = []
    for d in range(1, degree + 1):
        features.append(x ** d)

    X = np.hstack(features)

    # --- CAMBIO CLAVE: Normalización (Z-score scaling) ---
    # Esto evita que x^9 rompa el cálculo
    X_mean = np.mean(X, axis=0)
    X_std = np.std(X, axis=0)
    # Evitar división por cero
    X_std[X_std == 0] = 1
    X_scaled = (X - X_mean) / X_std

    return X_scaled

def loss_elastic_net(y, y_pred, w, lambda_, rho):
    mse = np.mean((y - y_pred)**2) / 2
    l1 = rho * lambda_ * np.sum(np.abs(w))
    l2 = (1 - rho) * lambda_ * np.sum(w**2)
    return mse + l1 + l2

def regression_elastic_net(x, y, degree=1, lr=0.001, lambda_=0.1, rho=0.5, max_iters=1000):
    X = build_features(x, degree)
    w = np.random.normal(0, 0.1, X.shape[1]) # Inicialización más pequeña
    b = 0.0
    losses = []

    for _ in range(max_iters):
        y_pred = X @ w + b

        # Gradientes
        # Nota: Ajustamos los términos de regularización para que coincidan con la pérdida
        dw = (
            -X.T @ (y - y_pred) / len(y) # Gradiente del MSE
            + rho * lambda_ * np.sign(w) # Lasso (L1)
            + 2 * (1 - rho) * lambda_ * w # Ridge (L2)
        )
        db = -np.mean(y - y_pred)

        w -= lr * dw
        b -= lr * db

        loss = loss_elastic_net(y, y_pred, w, lambda_, rho)
        losses.append(loss)

        # Si el loss es NaN, algo explotó
        if np.isnan(loss):
            break

    return w, b, losses

# --- Ejecución ---
np.random.seed(42)
x = np.linspace(-3, 3, 120)
y_target = np.sin(x) + np.random.normal(0, 0.2, size=len(x))

# Probamos con Grado 9
degrees_to_test = [1, 3, 5, 9]

plt.figure(figsize=(12, 8))
plt.scatter(x, y_target, color='gray', alpha=0.5, label="Datos reales")

for degree in degrees_to_test:
    # Usamos un lr más pequeño para estabilidad
    w, b, losses = regression_elastic_net(x, y_target, degree=degree, lr=0.01, lambda_=0.01, rho=0.5, max_iters=2000)

    X_for_pred = build_features(x, degree)
    y_pred = X_for_pred @ w + b

    plt.plot(x, y_pred, label=f"Grado {degree}")
    print(f"Grado {degree} completado. Bias: {b:.4f}")

plt.legend()
plt.title("Regresión Elastic Net con Polinomios de Alto Grado")
plt.grid(True)
plt.show()
```

```python
%%manim -ql NonLinearRegression

def get_regression_params(x, y, degree):
    x_input = np.atleast_2d(x).T if x.ndim == 1 else x
    features = [x_input ** d for d in range(1, degree + 1)]
    X = np.hstack(features)
    X_mean, X_std = np.mean(X, axis=0), np.std(X, axis=0)
    X_std[X_std == 0] = 1
    X_scaled = (X - X_mean) / X_std
    w, b = np.zeros(X.shape[1]), 0.0
    lr = 0.1
    for _ in range(5000):
        y_pred = X_scaled @ w + b
        error = y_pred - y
        dw = (X_scaled.T @ error) / len(y) + 0.01 * np.sign(w)
        db = np.mean(error)
        w -= lr * dw
        b -= lr * db
    return w, b, X_mean, X_std

class NonLinearRegression(Scene):
    def construct(self):
        title = Text("Regresión polinomial", font_size=36).to_edge(UP)
        axes = Axes(
            x_range=[-3.5, 4.5, 1],
            y_range=[-4, 4, 1],
            x_length=9, y_length=5,
            axis_config={"include_tip": False}
        ).shift(DOWN * 0.3)
        self.add(title, axes)

        np.random.seed(42)
        x_data = np.linspace(-3, 3, 65)
        y_data = np.sin(x_data) * 2 + np.random.normal(0, 0.8, size=len(x_data))
        dots = VGroup(*[Dot(axes.c2p(cx, cy), radius=0.04, color=BLUE_B, fill_opacity=0.7) for cx, cy in zip(x_data, y_data)])
        self.play(LaggedStart(*[FadeIn(d) for d in dots], lag_ratio=0.03, run_time=1.5))

        grados_interes = [1, 2, 3, 39, 43]
        curves, labels, formula_mobjects = [], [], []

        for deg in grados_interes:
            w, b, mu, sigma = get_regression_params(x_data, y_data, deg)
            def make_predict(w_val, b_val, mu_val, sigma_val, d_val):
                return lambda x_val: ((((np.array([x_val**i for i in range(1, d_val + 1)]) - mu_val) / sigma_val) @ w_val) + b_val)

            current_predict = make_predict(w, b, mu, sigma, deg)
            color = RED if deg <= 2 else (GREEN if deg <= 6 else ORANGE)
            new_curve = axes.plot(current_predict, x_range=[-3, 3], color=color)
            new_label = Text(f"Grado: {deg}", font_size=20, color=color).next_to(axes.c2p(3, current_predict(3)), UR, buff=0.1)

            # Construcción robusta de la cadena LaTeX
            if deg == 1:
                raw_formula = r"\hat{y} = w_1x + b"
            elif deg == 2:
                raw_formula = r"\hat{y} = w_2x^2 + w_1x + b"
            elif deg == 3:
                raw_formula = r"\hat{y} = w_3x^3 + w_2x^2 + w_1x + b"
            else:
                raw_formula = r"\hat{y} = w_{" + str(deg) + r"}x^{" + str(deg) + r"} + \dots + b"

            new_formula = MathTex(raw_formula, font_size=22, color=color)

            # Cuadro ajustado a la esquina inferior de la GRÁFICA (axes)
            f_rect = RoundedRectangle(
                corner_radius=0.1, fill_color=BLACK, fill_opacity=0.8, stroke_color=color,
                width=new_formula.width + 0.4, height=new_formula.height + 0.3
            )
            # Agrupamos y posicionamos relativo a la esquina inferior derecha de los EJES
            formula_group = VGroup(f_rect, new_formula).move_to(axes.get_corner(DR)).shift(LEFT*1.2 + UP*0.5)

            curves.append(new_curve); labels.append(new_label); formula_mobjects.append(formula_group)

        curr_curve, curr_label, curr_formula = curves[0], labels[0], formula_mobjects[0]
        self.play(Create(curr_curve), Write(curr_label), FadeIn(curr_formula))

        # 1. Underfitting
        u_txt = Text("Underfitting: El modelo es demasiado\nsencillo para captar el patrón.", font_size=22)
        u_rect = RoundedRectangle(corner_radius=0.1, fill_color=BLACK, fill_opacity=0.8, stroke_color=RED,
                                width=u_txt.width + 0.4, height=u_txt.height + 0.3)
        u_box = VGroup(u_rect, u_txt).next_to(axes, UP, buff=-1.2).shift(LEFT * 1.5)
        self.play(FadeIn(u_box, shift=UP), Indicate(curr_curve, color=RED))
        self.wait(2); self.play(FadeOut(u_box))

        # 2. Explicación
        p_txt = Text("La regresión polinomial nos permite modelar relaciones\nno lineales al incorporar términos de mayor grado,\nlogrando así un mejor ajuste.", font_size=20)
        p_rect = RoundedRectangle(corner_radius=0.1, fill_color=BLACK, fill_opacity=0.8, stroke_color=GREEN,
                                width=p_txt.width + 0.4, height=p_txt.height + 0.3)
        p_box = VGroup(p_rect, p_txt).move_to(u_box)
        self.play(FadeIn(p_box, shift=DOWN)); self.wait(4); self.play(FadeOut(p_box))

        # Ciclo de transformación (Actualiza curva, etiqueta y fórmula)
        for i in range(1, len(curves)):
            if grados_interes[i] == 39:
                new_title = Text("¡Overfitting detectado!", color=ORANGE, font_size=36).to_edge(UP)
                self.play(Transform(title, new_title))

            self.play(
                Transform(curr_curve, curves[i]),
                Transform(curr_label, labels[i]),
                Transform(curr_formula, formula_mobjects[i]),
                run_time=3.5 if grados_interes[i] >= 39 else 1.2
            )
            self.wait(1)

        self.play(Write(Text("El modelo memoriza el ruido, no el patrón.", font_size=24, color=ORANGE).to_edge(DOWN)))
        self.wait(2)
```

```python
%%manim -ql RidgeLassoElasticDemo


class RidgeLassoElasticDemo(Scene):
    def construct(self):
        # =============================================================
        # 0. MODELACION DE PROBLEMA
        # =============================================================
        problem_title = Text("Problema: Modelar funciones no lineales", color=BLUE, font_size=36).to_edge(UP)
        self.play(Write(problem_title))

        # Se enseña el modelo cuadratico
        quadratic_model = MathTex(
            r"f(x) = ax^2 + bx + c",
            color=GREEN
        ).scale(1.2)
        self.play(Write(quadratic_model))
        self.wait(1.5)

        # Función de pérdida
        loss_function = MathTex(
            r"f(a,b,c) = \frac{1}{2m}\sum_{i=1}^{m}\left(y^{(i)} - (ax^{(i)2} + bx^{(i)} + c)\right)^2"
        ).scale(0.85).next_to(quadratic_model, DOWN, buff=0.8)
        self.play(ReplacementTransform(quadratic_model, loss_function))
        self.wait(2)

        # Cálculo de gradientes individuales para cada coef.
        gradient_title = Text("Cálculo de gradientes individuales", 
                            color=RED, font_size=32).to_edge(UP)
        self.play(ReplacementTransform(problem_title, gradient_title))
        self.wait(0.5)

        gradient_group = VGroup()
        # Gradiente para 'a'
        grad_a = MathTex(
            r"\frac{\partial f}{\partial a} = -\frac{1}{m}\sum_{i=1}^{m}\left(y^{(i)} - (ax^{(i)2} + bx^{(i)} + c)\right)x^{(i)2}"
        ).scale(0.55)
        gradient_group.add(grad_a)

        # Gradiente para 'b'
        grad_b = MathTex(
            r"\frac{\partial f}{\partial b} = -\frac{1}{m}\sum_{i=1}^{m}\left(y^{(i)} - (ax^{(i)2} + bx^{(i)} + c)\right)x^{(i)}"
        ).scale(0.55)
        gradient_group.add(grad_b)
        
        # Gradiente para 'c'
        grad_c = MathTex(
            r"\frac{\partial f}{\partial c} = -\frac{1}{m}\sum_{i=1}^{m}\left(y^{(i)} - (ax^{(i)2} + bx^{(i)} + c)\right)"
        ).scale(0.55)
        gradient_group.add(grad_c)
        
        gradient_group.arrange(DOWN, buff=0.3).move_to(UP*0.5)

        # Mostramos cada gradiente
        self.play(Write(gradient_group))
        self.wait(1.5)

        # Mensaje de tedioso
        tedioso = Text("¡Demasiado tedioso para polinomios de alto grado!", 
                      color=YELLOW, font_size=28).to_edge(DOWN, buff=0.8)
        self.play(Write(tedioso))
        self.wait(1.5)

        # Eliminación de todos los elementos de esta sección
        self.play(FadeOut(gradient_title, grad_a, grad_b, grad_c, tedioso, loss_function), run_time=1.0)
        self.wait(0.5)  # Pausa breve para asegurar transición limpia

        # =============================================================
        # 1. GENERALIZACION MATRICIAL (Vandermonde + Gradiente Vectorial)
        # =============================================================
        matrix_title = Text("Solución: Formulación Matricial", color=BLUE, font_size=28).to_edge(UP)
        self.play(ReplacementTransform(gradient_title, matrix_title))
        
        # Explicación conceptual breve
        explicacion = Text(
            "Al modificar la matriz X con potencias (Features),\n"
            "el problema se vuelve lineal respecto a los parámetros θ.",
            font_size=20, line_spacing=1.2, color=GRAY
        ).next_to(matrix_title, DOWN, buff=0.3)

        self.play(Write(explicacion))
        self.wait(1.5)

        # Modelo matricial: f(x) = XΘ
        model_mat = MathTex(
            r"\hat{Y} = X \Theta",
            color=GREEN
        ).scale(1.2).next_to(explicacion, DOWN, buff=0.8)
        self.play(Write(model_mat))
        self.wait(1)

               # Definir X, Θ, Y con subíndices en amarillo (forma segura)
        X_main = MathTex(r"X = \begin{bmatrix} x_1^2 & x_1 & 1 \\ \vdots & \vdots & \vdots \\ x_m^2 & x_m & 1 \end{bmatrix}")
        X_sub = MathTex(r"m \times 3", color=YELLOW)
        X_def = VGroup(X_main, X_sub)
        X_sub.next_to(X_main, DOWN, buff=0.1).shift(RIGHT * 0.3)

        Theta_main = MathTex(r"\Theta = \begin{bmatrix} a \\ b \\ c \end{bmatrix}")
        Theta_sub = MathTex(r"3 \times 1", color=YELLOW)
        Theta_def = VGroup(Theta_main, Theta_sub)
        Theta_sub.next_to(Theta_main, DOWN, buff=0.1).shift(RIGHT * 0.2)

        Y_main = MathTex(r"Y = \begin{bmatrix} y^{(1)} \\ \vdots \\ y^{(m)} \end{bmatrix}")
        Y_sub = MathTex(r"m \times 1", color=YELLOW)
        Y_def = VGroup(Y_main, Y_sub)
        Y_sub.next_to(Y_main, DOWN, buff=0.1).shift(RIGHT * 0.2)

        definitions = VGroup(X_def, Theta_def, Y_def).scale(0.7).arrange(RIGHT, buff=0.8).next_to(model_mat, DOWN, buff=0.8)
        self.play(FadeIn(definitions))
        self.wait(2)

        # Función de pérdida en notación matricial
        loss_vector = MathTex(
            r"L(\Theta) = \frac{1}{2m} \| X\Theta - Y \|^2",
            color=RED
        ).scale(0.9).next_to(definitions, DOWN, buff=0.8)
        self.play(Write(loss_vector))
        self.wait(2)

        # Gradiente vectorial
        grad_vector = MathTex(
            r"\nabla_\Theta L = \begin{bmatrix} \dfrac{\partial L}{\partial a} \\[0.8em] \dfrac{\partial L}{\partial b} \\[0.8em] \dfrac{\partial L}{\partial c} \end{bmatrix}",
            color=PURPLE
        ).scale(0.8).next_to(loss_vector, DOWN, buff=0.8)

        self.play(Write(grad_vector))
        self.wait(1.5)

        # Desglose de cada derivada (igual que en la sección escalar, pero alineado)
        # Usamos RIGHT para colocarlas a la derecha del gradiente
        grad_a_exp = MathTex(
            r"= -\frac{1}{m}\sum_{i=1}^{m}\left(y^{(i)} - (a x^{(i)2} + b x^{(i)} + c)\right)x^{(i)2}"
        ).scale(0.55).next_to(grad_vector[0][1], RIGHT, buff=0.3)  # [0][1] es el primer \partial L / \partial a

        grad_b_exp = MathTex(
            r"= -\frac{1}{m}\sum_{i=1}^{m}\left(y^{(i)} - (a x^{(i)2} + b x^{(i)} + c)\right)x^{(i)}"
        ).scale(0.55).next_to(grad_vector[0][3], RIGHT, buff=0.3)  # [0][3] es el segundo término

        grad_c_exp = MathTex(
            r"= -\frac{1}{m}\sum_{i=1}^{m}\left(y^{(i)} - (a x^{(i)2} + b x^{(i)} + c)\right)"
        ).scale(0.55).next_to(grad_vector[0][5], RIGHT, buff=0.3)  # [0][5] es el tercer término

        # Animar los desgloses
        self.play(Write(grad_a_exp))
        self.play(Write(grad_b_exp))
        self.play(Write(grad_c_exp))
        self.wait(2)

        self.play(FadeOut(loss_vector))
        # Ventaja clave
        advantage = Text(
            "¡Ahora podemos calcular todas las derivadas\n"
            "simultáneamente con una sola expresión matricial!",
            color=GREEN, font_size=24
        ).to_edge(DOWN)
        self.play(Write(advantage))
        self.wait(2)

        # Fade out todo antes de la transición
        self.play(
            FadeOut(matrix_title, explicacion, model_mat, definitions, loss_vector, grad_vector, grad_a_exp, grad_b_exp, grad_c_exp, advantage),
            run_time=1.0
        )
        self.wait(0.5)

        # Transición a las regularizaciones
        transition_text = Text(
            "Pero ¿qué pasa cuando tenemos muchos parámetros?\n"
            "¡Necesitamos regularización para evitar el sobreajuste!",
            color=YELLOW, font_size=28
        )
        self.play(Write(transition_text))
        self.wait(3)
        self.play(FadeOut(transition_text), run_time=1.0)
        self.wait(0.5)  # Pausa breve para asegurar transición limpia
        # =============================================================
        # 2. CONFIGURACIÓN GLOBAL (EJES Y PARÁMETROS)
        # =============================================================
        t_lasso = 1.0   # Para el rombo de Lasso
        r_ridge = 1.5   # Para el círculo de Ridge
        
        ejes = Axes(
            x_range=[-2, 2, 1],
            y_range=[-2, 2, 1],
            axis_config={"include_tip": True, "stroke_width": 2},
            x_length=6,
            y_length=6
        ).to_edge(RIGHT)
        
        labels = ejes.get_axis_labels(r"w_1", r"w_2")
        self.add(ejes, labels)

        # =============================================================
        # 3. LASSO (L1)
        # =============================================================
        tit_l1 = Text("1. Lasso (L1): Selección", color=GREEN, font_size=32).to_edge(UL, buff=0.5)
        form_l1 = MathTex(r"|w_1| + |w_2| \leq t", color=GREEN, font_size=28).next_to(tit_l1, DOWN, aligned_edge=LEFT)
        
        # Título para la fórmula principal (con penalización)
        title_loss_penalty_l1 = Text("Pérdida Mínima + Penalización", color=YELLOW).scale(0.5).next_to(form_l1, DOWN, buff=0.5).align_to(tit_l1, LEFT)
        
        # Añadir la función de pérdida y regularización
        loss_equation = MathTex(
            r"\frac{1}{2N}\sum_{i=1}^{N}\left(y^{(i)} - \hat{y}^{(i)}\right)^2 + \lambda\|w\|_1"
        ).scale(0.6).next_to(title_loss_penalty_l1, DOWN, buff=0.3).align_to(tit_l1, LEFT)

        # Título para la fórmula secundaria
        title_loss_only_l1 = Text("Pérdida Mínima:", color=RED).scale(0.5).next_to(loss_equation, DOWN, buff=0.5).align_to(tit_l1, LEFT)

        # CUADRO DE PÉRDIDA MÍNIMA - POSICIONADO EN LA PARTE INFERIOR IZQUIERDA
        min_loss_box_l1 = VGroup(
        MathTex(r"\frac{1}{2N}\sum_{i=1}^{N}\left(y^{(i)} - \hat{y}^{(i)}\right)^2"),
        MathTex(r"\text{donde } \hat{y}^{(i)} = w_1x_1^{(i)} + w_2x_2^{(i)}")
        ).scale(0.6).arrange(DOWN, buff=0.2).next_to(title_loss_only_l1, DOWN, buff=0.3).align_to(tit_l1, LEFT)

        diamante = Polygon(
            ejes.c2p(t_lasso, 0), 
            ejes.c2p(0, t_lasso), 
            ejes.c2p(-t_lasso, 0), 
            ejes.c2p(0, -t_lasso),
            color=GREEN,
            fill_opacity=0.3,
            stroke_width=4
        )

        beta_hat = Dot(ejes.c2p(0.82, 1), color=RED)
        txt_beta = MathTex(r"\hat{\beta}", color=RED, font_size=30).next_to(beta_hat, UR, buff=0.1)

        elipses = VGroup()
        for s in [0.3, 0.5, 0.9, 1.2]:
            e = Ellipse(
                width=2.2 * s, 
                height=1.1 * s, 
                color=RED, 
                stroke_opacity=0.4
            ).rotate(PI / 15).move_to(beta_hat.get_center())
            elipses.add(e)

        punto_choque = Dot(ejes.c2p(0, t_lasso), color=YELLOW, radius=0.1)
        txt_sparsity = Text("w₁ = 0 (Sparsity)", color=YELLOW, font_size=22, weight=BOLD).next_to(punto_choque, LEFT, buff=0.2)

        self.play(Write(tit_l1), Write(form_l1))
        self.play(Write(title_loss_penalty_l1))
        self.play(Write(loss_equation))
        self.play(Write(title_loss_only_l1))
        self.play(Write(min_loss_box_l1))
        self.play(FadeIn(beta_hat, txt_beta))
        self.play(Create(diamante))
        self.play(LaggedStart(*[Create(e) for e in elipses], lag_ratio=0.4, run_time=3))
        self.play(FadeIn(punto_choque), Write(txt_sparsity))
        self.play(Indicate(punto_choque, color=YELLOW))
        self.wait(2)

        # TRANSICIÓN LIMPIA A RIDGE
        self.play(
            FadeOut(tit_l1, form_l1, title_loss_penalty_l1, loss_equation, title_loss_only_l1, min_loss_box_l1, diamante, elipses, punto_choque, txt_sparsity, loss_equation),
            run_time=1.0
        )
        self.wait(0.5)

        # =============================================================
        # 4. RIDGE (L2)
        # =============================================================
        tit_l2 = Text("2. Ridge (L2): Suavizado", color=BLUE, font_size=32).to_edge(UL, buff=0.5)
        form_l2 = MathTex(r"w_1^2 + w_2^2 \leq t", color=BLUE, font_size=28).next_to(tit_l2, DOWN, aligned_edge=LEFT)
        # Título para la fórmula principal (con penalización)
        title_loss_penalty = Text("Pérdida Mínima + Penalización", color=YELLOW).scale(0.5).next_to(form_l2, DOWN, buff=0.5).align_to(tit_l2, LEFT)

        # Añadir la función de pérdida y regularización
        loss_equation_ridge = MathTex(
            r"\frac{1}{2N}\sum_{i=1}^{N}\left(y^{(i)} - \hat{y}^{(i)}\right)^2 + \lambda\|w\|_2^2"
        ).scale(0.6).next_to(form_l2, DOWN, buff=1.5).align_to(tit_l2, LEFT)
        
        # Título para la fórmula secundaria
        title_loss_only = Text("Pérdida Mínima:", color=RED).scale(0.5).next_to(loss_equation_ridge, DOWN, buff=0.5).align_to(tit_l2, LEFT)

        # CUADRO DE PÉRDIDA MÍNIMA A LA IZQUIERDA
        min_loss_box = VGroup(
            MathTex(r"\frac{1}{2N}\sum_{i=1}^{N}\left(y^{(i)} - \hat{y}^{(i)}\right)^2"),
            MathTex(r"\text{donde } \hat{y}^{(i)} = w_1x_1^{(i)} + w_2x_2^{(i)}")
        ).scale(0.6).arrange(DOWN, buff=0.2).to_edge(DOWN, buff=1.5).align_to(tit_l2, LEFT)

        circulo = Circle(
            radius=r_ridge,
            color=BLUE,
            fill_opacity=0.25,
            stroke_width=4
        ).move_to(ejes.c2p(0, 0))

        punto_tangencia = Dot(
            ejes.c2p(0.5, 0.85),
            color=YELLOW,
            radius=0.1
        )
        txt_tangencia = Text("Solución Ridge\n(tangencia circular)", color=YELLOW, font_size=18).next_to(punto_tangencia, LEFT, buff=0.3)

        self.play(Write(tit_l2), Write(form_l2))
        self.play(Write(title_loss_penalty))
        self.play(Write(loss_equation_ridge))
        self.play(Write(title_loss_only))
        self.play(Create(circulo))
        self.play(Write(min_loss_box))
        self.play(LaggedStart(*[Create(e) for e in elipses], lag_ratio=0.4, run_time=3))
        self.play(FadeIn(punto_tangencia), Write(txt_tangencia))
        self.play(Indicate(punto_tangencia, color=YELLOW))
        self.wait(4)

        # TRANSICIÓN LIMPIA A ELASTIC NET
        self.play(
            FadeOut(tit_l2, form_l2, title_loss_penalty ,loss_equation_ridge, title_loss_only, min_loss_box, circulo, punto_tangencia, txt_tangencia,beta_hat, txt_beta, elipses),
            run_time=1.0
        )
        self.wait(0.5)

        # =============================================================
        # 5. ELASTIC NET
        # =============================================================
        tit_en = Text("3. Elastic Net: Híbrido", color=PURPLE, font_size=32).to_edge(UL, buff=0.5)
        # FÓRMULA DE ELASTIC NET A LA IZQUIERDA
        title_loss_penalty_en = Text("Minimize loss + penalty", color=YELLOW).scale(0.5).next_to(tit_en, DOWN, buff=0.5).align_to(tit_en, LEFT)
       
        # Fórmula principal de Elastic Net
        loss_equation_en = MathTex(
            r"\frac{1}{N}\sum_{i=1}^{N}\left(y^{(i)} - \hat{y}^{(i)}\right)^2 + \alpha\left(\lambda\|w\|_1 + \frac{1-\lambda}{2}\|w\|_2^2\right)"
        ).scale(0.6).next_to(title_loss_penalty_en, DOWN, buff=0.3).align_to(tit_en, LEFT)
        
        # Condición (donde 0 < α < 1 y 0 < λ < 1)
        condition_en = MathTex(
            r"\text{donde } 0 < \alpha < 1 \text{ y } 0 < \lambda < 1"
        ).scale(0.6).next_to(loss_equation_en, DOWN, buff=0.3).align_to(tit_en, LEFT)
        
        origin = ejes.c2p(0,0)

        # REFERENCIAS CON TUS MEDIDAS EXACTAS
        l2_dash = DashedVMobject(
            Circle(radius=r_ridge, color=BLUE),
            num_dashes=30
        ).move_to(origin)
        
        l1_dash = DashedVMobject(
            Polygon(
                ejes.c2p(t_lasso, 0), 
                ejes.c2p(0, t_lasso), 
                ejes.c2p(-t_lasso, 0), 
                ejes.c2p(0, -t_lasso),
                color=GREEN
            ),
            num_dashes=30
        ).move_to(origin)
        
        def superellipse(u, n=1.5):
            """Superelipse para Elastic Net (n=1.5)"""
            x = np.sign(np.cos(u)) * np.abs(np.cos(u))**(2/n)
            y = np.sign(np.sin(u)) * np.abs(np.sin(u))**(2/n)
            return np.array([x, y, 0])
        
    
        
        # ELASTIC NET: ROMBO CON ESQUINAS REDONDEADAS
        elastic_shape = ParametricFunction(
            superellipse,
            t_range=[0, 2*PI],
            color=PURPLE,
            fill_opacity=0.4,
            stroke_width=4
        ).scale(t_lasso).move_to(origin)

        # Ridge: línea punteada azul
        ridge_sample = DashedLine(ORIGIN, RIGHT * 0.8, color=BLUE, dash_length=0.1)
        
        # Lasso: línea de puntos verdes
        lasso_sample = VGroup()
        num_dots = 15
        for i in range(num_dots):
            dot = Dot(
                ORIGIN + RIGHT * (0.8 * i / (num_dots - 1)),
                color=GREEN,
                radius=0.03
            )
            lasso_sample.add(dot)
        
        # Elastic Net: línea sólida morada
        elastic_sample = Line(ORIGIN, RIGHT * 0.8, color=PURPLE, stroke_width=3)
        
        # Construccion de la leyenda
        leyenda_title = Text("Donde:", color=YELLOW).scale(0.5).align_to(condition_en, LEFT)
        
        ridge_label = VGroup(ridge_sample, Text("Ridge", font_size=22)).arrange(RIGHT, buff=0.15)
        lasso_label = VGroup(lasso_sample, Text("Lasso", font_size=22)).arrange(RIGHT, buff=0.15)
        elastic_label = VGroup(elastic_sample, Text("Elastic Net", font_size=22)).arrange(RIGHT, buff=0.15)
        
        # Alineamos la leyenda con el texto "donde 0 < α < 1..."
        leyenda = VGroup(leyenda_title, ridge_label, lasso_label, elastic_label).arrange(
            DOWN, 
            aligned_edge=LEFT, 
            buff=0.25
        ).next_to(condition_en, DOWN, buff=0.5)

        # ANIMACIÓN FINAL
        self.play(Write(tit_en))
        self.play(Write(title_loss_penalty_en))
        self.play(Write(loss_equation_en))
        self.play(Write(condition_en))
        self.play(Create(l2_dash), Create(l1_dash), run_time=1.5)
        self.play(Create(elastic_shape), run_time=1.2)
        self.play(FadeIn(leyenda))
        self.wait(4)
        
        # DESTACAR RELACIÓN CLAVE
        self.play(
            Indicate(l2_dash, color=BLUE, scale_factor=1.05),
            Indicate(l1_dash, color=GREEN, scale_factor=1.05),
            run_time=1.5
        )
        self.wait(4)
```
